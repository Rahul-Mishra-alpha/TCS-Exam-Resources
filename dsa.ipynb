```python
#Print all prime numbers from 1 to N.
dt=[]
for i in range(2,num+1):
    f=1
    for j in range(2,i//2+1):
        if i%j==0:
            f=0
            break
    if f==1:
        dt.append(i)
print(dt)
```

```python
#Strong number check (sum of factorial of digits = the number).
a=145
s=str(a)
def fact(i):
    p=1
    for j in range(1,i+1):
        p=p*j
    return p
sm=0
for i in s:
    sm=sm+fact(int(i))

if sm==a:
    print(a)
else:
    print("not so")
    
```

```python
#Harshad / Niven number check (number divisible by sum of its digits).
hn=18
sh=str(hn)
summ=0
for i in sh:
    summ=summ+int(i)


if hn%summ==0:
    print(hn)
else:
    print("lol")
```

```python
#Armstrong number check.
an=153
sh=str(an)
summ=0
for i in sh:
    summ=summ+(int(i)**3)


if an==summ:
    print(an)
else:
    print("lol")
```
